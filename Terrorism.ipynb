Yes, you can visualize the unique values present in the data. For this, we will be using the seaborn library. You have to call the sns.countlot() function and specify the variable to plot the count plot.

## Affected regions due to terrorist attacks

```python
# assuming 'df' is the pandas DataFrame
region_count = df['Region'].value_counts()

# create bar plot using Seaborn
sns.set_style('whitegrid')
plt.figure(figsize=(10,5))
ax = sns.barplot(x=region_count.index, y=region_count.values, palette='muted')
ax.set_title('Terrorist Activities by Region')
ax.set_xlabel('Region')
ax.set_ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Number of killed  and wonded over years

```python
# Create a pivot table to calculate the sum of people killed and wounded for each year
year_wise = df.pivot_table(values=['Killed', 'Wounded'], index=['Year'], aggfunc='sum')

# Create a line plot to visualize the number of people killed and wounded over the years
plt.plot(year_wise.index, year_wise['Killed'], label='Killed')
plt.plot(year_wise.index, year_wise['Wounded'], label='Wounded')

# Add title and axis labels
plt.title('Number of People Killed and Wounded Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of People')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

## Year-wise increase in terror activities in regions

```python
# Create a new DataFrame with the count of terror activities per year and region
df2 = pd.DataFrame(df.groupby(['Year', 'Region']).size(), columns=['Count']).reset_index()

# Pivot the DataFrame to have the years as the index and the regions as columns
df3 = df2.pivot(index='Year', columns='Region', values='Count').fillna(0)

# Calculate the year-wise increase in terror activities for each region
df4 = df3.diff()


# Create a line chart of the year-wise increase in terror activities by region
df4.plot(kind='line',figsize=(18,6))

# Add a title and axis labels
plt.title('Year-wise Increase in Terror Activities by Region')
plt.xlabel('Year')
plt.ylabel('Increase in Terror Activities')

# Move the legend to the top
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=len(df4.columns))

# Display the plot
plt.show()
```

```python
# Create a new DataFrame with the count of terrorist activities per year and group
df2 = pd.DataFrame(df.groupby(['Year', 'Group']).size(), columns=['Count']).reset_index()

# Pivot the DataFrame to have the years as the index and the groups as columns
df3 = df2.pivot(index='Year', columns='Group', values='Count').fillna(0)

# Calculate the year-wise increase in terrorist groups
df4 = df3.diff()

# Get the top 10 terrorist groups based on the total number of activities
top_groups = df.groupby('Group')['Group'].count().sort_values(ascending=False).head(10).index.tolist()

# Create a line chart of the year-wise increase in terrorist groups
df4[top_groups].plot(kind='line', figsize=(18,6))

# Add a title and axis labels
plt.title('Year-wise Increase in Terrorist Groups')
plt.xlabel('Year')
plt.ylabel('Increase in Terrorist Groups')

# Move the legend to the top
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=len(df4.columns))

# Display the plot
plt.show()
```
